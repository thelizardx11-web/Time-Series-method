```python
# import libabries 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create a sample time serires dataset 
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', end='2023-12-31')
sales = np.random.randint(200, 1000, size=len(dates)) + np.sin(np.linspace(0, 20, len(dates))) * 1 

df= pd.DataFrame({'Date': dates, 'Sales': sales})
print("Sample Time Series Data Created Sucessfully!\n")
print(df.head())
```

```python
# Basic information
print("\nDataset Info:")
print(df.head())
```

```python
# missing values 
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Set date as index 
df.set_index('Date', inplace=True)
```

```python
# Basic summary statstics 
print("\nSummary Statisctics:")
print(df.describe())
```

```python
# Visulazation - Trend + disturbition 
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.lineplot(x=df.index, y=df['Sales'], ax=axes[0], color='blue')
axes[0].set_title("Sales Trend Over Time")

sns.histplot(df['Sales'], bins=30, ax=axes[1], kde=True, color='orange')
axes[1].set_title("Sales Distribution")
plt.show()
```

```python
# Correlation & Boxplot Example
# (Assume we add another variable like 'Marketing Spend')
df['Marketing_Spend'] = df['Sales'] * np.random.uniform(0.2, 0.5, size=len(df))

corr = df.corr(numeric_only=True)
print("\n Correlation Matrix:\n", corr)

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='viridis')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# boxplot 
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Sales'], color='coral')
plt.title("Boxplot of Sales (Check Outliers)")
plt.show()

print("\n EDA Completed Successfully  No Errors, All Good!")
```
